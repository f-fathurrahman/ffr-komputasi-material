$$
\begin{equation}
\omega = 2 \sqrt{\frac{\kappa}{m}}
\left|
\sin\left(
\frac{ka}{2}
\right)
\right|
\end{equation}
$$

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from math import fabs, sin, sqrt, pi
```

```python
a = 1.0
Npoints = 500
κ = 1.0
m = 1.0
k = np.linspace(-2*pi/a,2*pi/a,Npoints)
ω = np.zeros(Npoints)
```

```python
for i in range(Npoints):
    ω[i] = 2*sqrt(κ/m)*fabs(sin(k[i]*a/2))
```

```python
%matplotlib inline
matplotlib.style.use("default")
plt.clf()
plt.grid(True)
plt.plot(k, ω)
```
