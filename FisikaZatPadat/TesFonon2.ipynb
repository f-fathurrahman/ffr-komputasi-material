```python
from math import sqrt, cos
```

```python
a = 1.0
def calc_ω1(κ1, κ2, m, k):
    return sqrt( (κ1 + κ2)/m + sqrt( κ1**2 + κ2**2 + 2*κ1*κ2*cos(k*a))/m )
    
def calc_ω2(κ1, κ2, m, k):
    return sqrt( (κ1 + κ2)/m - sqrt( κ1**2 + κ2**2 + 2*κ1*κ2*cos(k*a))/m )
```

```python
κ1 = 1.0
κ2 = 1.5
m = 1.0
```

```python
print( calc_ω1(κ1, κ2, m, 0.0) )
print( calc_ω2(κ1, κ2, m, 0.0) )
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from math import pi
```

```python
Npoints = 201
k = np.linspace(-2*pi/a,2*pi/a,Npoints)
omega_akustik = np.zeros(Npoints)
omega_optik = np.zeros(Npoints)
for i in range(Npoints):
    omega_optik[i] = calc_ω1(κ1, κ2, m, k[i])
    omega_akustik[i] = calc_ω2(κ1, κ2, m, k[i])
```

```python
matplotlib.style.use("default")
plt.clf()
plt.figure(dpi=200)
plt.plot(k, omega_optik)
plt.plot(k, omega_akustik)
plt.grid(True)
```
