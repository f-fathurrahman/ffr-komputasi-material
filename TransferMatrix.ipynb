# Metode matriks transfer

Metode berdasarkan pada pencocokan fungsi gelombang (*wavefunction matching*). Misalkan kita memiliki potensial penghalang persegi:

$$
V(x) = \begin{cases}
V, \left|x\right| \leq a \\
0, \text{lainnya}
\end{cases}
$$

Untuk $E > V$ solusi dari keadaan hamburan memiliki bentuk:

$$
\Psi(x) = \begin{cases}
A_{L}\mathrm{e}^{\imath k x} + B_{L}\mathrm{e}^{-\imath k x},\,\, x < -a \\
A\mathrm{e}^{\imath k' x} + B\mathrm{e}^{-\imath k' x},\,\, -a \leq x \leq a \\
A_{R}\mathrm{e}^{\imath k x} + B_{R}\mathrm{e}^{-\imath k x},\,\, x > a
\end{cases}
$$

dengan:

$$
\begin{align}
k  & = \sqrt{\frac{2mE}{\hbar^2}} \\
k' & = \sqrt{\frac{2m(E-V)}{\hbar^2}}
\end{align}
$$

Dengan mensyaratkan kontinuitas dari fungsi gelombang dan turunan pertamanaya pada $x=-a$ dan $x=a$, kita memiliki 4 persamaan untuk 6 koefisien yang tidak diketahui. Dua persamaan lagi dari diturunkan dari syarat awal sehingga semuat koefisien dapat ditentukan.
